# plotting with matplotlib

```python
#import
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define potential energy
def pot_en(x, k=1.0): #predefined variables dont have to be recalled
    return 0.5 * k * x**2 #k spring constant
pot_en(1)
pot_en(2, 10)
```

```python
#define mesh for plotting
x = np.linspace(-5, 5, 30) # np.linspace(xmin, xmax, npts)
#define y values from x list
y = pot_en(x)
#test output
print(np.transpose([x, y]))
```

```python
#plot da function
plt.figure(figsize=(5,4)) #figure size in in.
plt.plot(x, y, color="red", linewidth=2)
#label thine axes
plt.xlabel("displacement x")
plt.ylabel("potential energy y")
plt.xlim(-4,4)
plt.ylim(0,12)

plt.show() #mix em together
```

## contour plots

```python
#import
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define function
def pe_2d(x, y, kx=1.0, ky=1.0): #predefined variables dont have to be recalled
    return 0.5 * (kx * x**2 + ky * y**2) #kx, ky spring constants

pe_2d(2, 1)
```

```python
#define meshes for plotting
x = np.linspace(-5, 5, 100) # np.linspace(xmin, xmax, npts)
y = np.linspace(-5, 5, 100)

#combino into meshgrido
X, Y = np.meshgrid(x, y)

#define Z
Z = pe_2d(X,Y)

#testo
#print(X, "\n", Y, "\n", Z)
```

```python
#plot
plt.figure(figsize=(5,5)) #figure size in in.
plt.contourf(X, Y, Z)

plt.colorbar(label="legend")

plt.xlabel("displacement")
plt.ylabel("potential energy")

plt.xlim(-5,5)
plt.ylim(-5,5)

plt.show() #mix em together
```
